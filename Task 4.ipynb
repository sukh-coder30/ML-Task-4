```python
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv("data.csv")
```

```python
#checking is there any null value
df.isnull().sum()
```

```python
#checking vaules in diagnosis
df['diagnosis'].value_counts()
```

```python
#encoding target variables

df['diagnosis']=df['diagnosis'].map({'B':0, 'M':1})
```

```python
df
```

```python
df=df.drop('id', axis=1)
```

```python
df
```

```python
#drop unnessary features

df= df.drop('Unnamed: 32',axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df.drop('diagnosis', axis=1)
y=df['diagnosis']
```

```python
x
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#scaling data

from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
x_train_scaled= scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
#train a classifier

from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```

```python
z= model.decision_function(x_test_scaled)
```

```python
z
```

```python
import numpy as np

sigmoid_probs = 1 / (1 + np.exp(-z))
```

```python
sigmoid_probs
```

```python
sorted_indices= np.argsort(z)
sorted_logist=z[sorted_indices]
sorted_prob=sigmoid_probs[sorted_indices]
plt.figure(figsize=(8,5))
plt.plot(sorted_logist, sorted_prob, label='sigmoid')
plt.axhline(0.5, color='red', linestyle='--', label='Threshold 0.5')
plt.xlabel('Logit (z)')
plt.ylabel('Probability')
plt.title('Sigmoid Function Applied to Breast Cancer Predictions')
plt.legend()
plt.grid()
plt.show()
```
